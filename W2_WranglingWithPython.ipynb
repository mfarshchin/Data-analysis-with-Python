## 1- PRE-PROCESSING DATA IN PYTHON

```python
# read csv file and import the data frame
import pandas as pd
import numpy as np
DF=pd.read_csv('CarsDF.csv')
DF.head()
```

```python
# simple data frame operations
# assign the data frame to a new variable: we should use pandas copy with deep=True
D1=DF.copy(deep=True)
D1['symboling']=D1['symboling']+1
D1.head()
DF.head()# note that the original data frame remains unchanged
```

```python
# remove a row of data frame (drop)
D1=DF.copy(deep=True)
# remove first row
D1.drop([0,1,3], inplace = True)
D1.head()
```

```python
# delete a column of data in a data frame
D1=DF.copy(deep=True)
# remove 'symboling' column
D1.drop(['symboling'], axis=1, inplace = True)
D1.head()
```

## 2- DEALING WITH MISSING VALUES IN PYTHON

- Delete the missing value
- Replace the missing value
- Leave it as missing value

```python
# to drop the ? values from the data frame rows, corresponing to the "normalized-losses" column
# we perform the following
D1=DF.copy(deep=True)
# replace the ? with NaN
D1 = D1.replace('?', np.NaN)
# use dropna function
D1=D1.dropna()
D1.head()
```

```python
# evaluating for missing data
D1=DF.copy(deep=True)
D1.replace("?", np.nan, inplace = True)
missing_data=D1.isnull()
missing_data.head(5)
```

```python
# count missing data in each column
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

## 3- DATA FORMATTING IN PYTHON

apply z-score scaling to the length column: $\frac{x-\mu_x}{\sigma_x}$

```python
D1['length']=(D1['length']-D1['length'].mean())/D1['length'].std()
D1['length']
```

```python
# convert data format
# first drop all rows with nan
D1.dropna(axis=0, inplace=True)
# Then reset index, because we droped rows
D1.reset_index(drop=True, inplace=True)
# now convert data format
df=D1
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
D1.dtypes
```
